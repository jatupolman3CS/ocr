```python
#!/usr/bin/python -tt 
# -*- coding: utf-8 -*-
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
import codecs
from sklearn.model_selection import train_test_split

def loadImageFolder(path): 
    print('please wait processing...')
    files = glob.glob(path+'*.png')
    filesize=len(files)
    print(filesize)
    data=[]
    label=[]
    for i,name in enumerate(files):
        try:
            pathIn, finame = os.path.split(name) #pathIn = pathfile, finame=file.xxx
            fi=os.path.splitext(finame)[0] #fi=filename without .xxx
            data.append(fi)
            fileGT=path+fi+'.gt.txt'
            f1 = codecs.open(fileGT, "r", "utf-8-sig")
            label.append(f1.read())
        except IOError as exc: #Not sure what error this is
            print('Error file is ', finame)
#            if exc.errno != errno.EISDIR:
#                raise
    print('#data=%d, #label=%d'%(len(data),len(label)))

    return data, label

def writeData(xdata,ydata,path,filename):
    pathname=path+filename+'.txt'
    with codecs.open(pathname,'w','utf-8-sig') as txtFile:
#        txtFile.write('Filename Transciption\n')
        for i,x in enumerate(xdata):
            print('%s %s' % (xdata[i], ydata[i]))
            txtFile.write('%s|%s\n' % (xdata[i], ydata[i]))
    txtFile.closed
    print('Write ',pathname,'complete...')

path='d:/hwocr/data/'
data, labels = loadImageFolder(path)
Xdata_train, Xdata_test, Ydata_train, Ydata_test = train_test_split(data, labels, test_size=0.1)
writeData(Xdata_train, Ydata_train,path, 'trainData')
writeData(Xdata_test, Ydata_test, path, 'testData')
```
